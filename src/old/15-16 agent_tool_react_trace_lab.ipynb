## 2) Tool ì •ì˜ (@tool)

- `@tool` ë°ì½”ë ˆì´í„°ë¡œ **LLMì´ ì‚¬ìš©í•  ìˆ˜ ìˆëŠ” í•¨ìˆ˜**ë¥¼ ë“±ë¡í•©ë‹ˆë‹¤.
- Docstring(ì„¤ëª…)ê³¼ íƒ€ì… íŒíŠ¸ê°€ **ì—ì´ì „íŠ¸ì˜ ì˜¬ë°”ë¥¸ ì‚¬ìš©**ì— í° ë„ì›€ì„ ì¤ë‹ˆë‹¤.

```python
from datetime import datetime, timezone, timedelta
from langchain_core.tools import tool

@tool
def add(a: int, b: int) -> int:
    """ë‘ ì •ìˆ˜ aì™€ bì˜ í•©ì„ ë°˜í™˜í•©ë‹ˆë‹¤."""
    return a + b

@tool
def multiply(a: int, b: int) -> int:
    """ë‘ ì •ìˆ˜ aì™€ bì˜ ê³±ì„ ë°˜í™˜í•©ë‹ˆë‹¤."""
    return a * b

@tool
def to_kst(iso_dt: str) -> str:
    """UTC ISO8601 ë‚ ì§œë¬¸ìì—´(ì˜ˆ:'2025-11-13T06:00:00Z')ì„ í•œêµ­ì‹œê°„(KST, UTC+9)ìœ¼ë¡œ ë³€í™˜í•˜ì—¬ ë°˜í™˜í•©ë‹ˆë‹¤."""
    dt = datetime.fromisoformat(iso_dt.replace('Z', '+00:00'))
    kst = dt.astimezone(timezone(timedelta(hours=9)))
    return kst.strftime('%Y-%m-%d %H:%M:%S KST')
```

## 3) ReAct Agent ìƒì„± (LangChain Hub í”„ë¡¬í”„íŠ¸)

- `hwchase17/react` í…œí”Œë¦¿ì„ LangChain Hubì—ì„œ ê°€ì ¸ì˜µë‹ˆë‹¤.
- `create_react_agent` + `AgentExecutor` ë¡œ ì—ì´ì „íŠ¸ë¥¼ êµ¬ì„±í•©ë‹ˆë‹¤.

```python
from langchain_openai import ChatOpenAI
from langgraph.prebuilt import create_react_agent
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage

llm = ChatOpenAI(model="gpt-4o-mini")
tools = [add, multiply, to_kst]

SYSTEM_PROMPT = (
    "ë„ˆëŠ” ê³„ì‚° ë° ì‹œê°„ ë³€í™˜ ë„êµ¬ë¥¼ í™œìš©í•´ ë‹¨ê³„ì ìœ¼ë¡œ ë¬¸ì œë¥¼ í‘¸ëŠ” ì¡°êµë‹¤. "
    "í•„ìš”í•˜ë©´ ë„êµ¬ë¥¼ ì‚¬ìš©í•˜ê³ , ìµœì¢…ì ìœ¼ë¡œ ê°„ê²°í•œ í•œêµ­ì–´ ë‹µë³€ì„ ì œì‹œí•˜ë¼. "
    "ë„êµ¬ ì…ë ¥ì€ JSON í˜•ì‹ì„ ìœ ì§€í•˜ë¼."
)

# ---- ë²„ì „ í˜¸í™˜ ìƒì„±ê¸° ----
def build_agent():
    # ìš°ì„  v2 + state_modifier ì‹œë„
    try:
        return create_react_agent(
            model=llm,
            tools=tools,
            state_modifier=SYSTEM_PROMPT,
            version="v2",          # ì¼ë¶€ ë²„ì „ì€ versionì„ ìš”êµ¬
        )
    except TypeError:
        pass

    # v1 + messages_modifier ì‹œë„
    try:
        return create_react_agent(
            model=llm,
            tools=tools,
            messages_modifier=SYSTEM_PROMPT,
            version="v1",
        )
    except TypeError:
        pass

    # ë§ˆì§€ë§‰ í´ë°±: modifier ì—†ì´ ìƒì„± â†’ í˜¸ì¶œ ì‹œ SystemMessageë¥¼ í•¨ê»˜ ë„˜ê¹€
    return create_react_agent(
        model=llm,
        tools=tools,
        # version ì¸ìë„ ì–´ë–¤ ì„¤ì¹˜ë³¸ì€ í•„ìˆ˜ì¼ ìˆ˜ ìˆìŒ â€” í•„ìš”í•˜ë©´ í•œ ë²ˆ ë” ì‹œë„
    )

agent = build_agent()

# === ë‹¨ê±´ ì‹¤í–‰ (modifierë¥¼ ëª» ë„£ì€ í´ë°± ì¼€ì´ìŠ¤ë„ ì»¤ë²„) ===
def run_once(question: str):
    try:
        # modifierê°€ ë¨¹íŒ ë²„ì „
        res = agent.invoke(
            {"messages": [HumanMessage(content=question)]},
            config={"recursion_limit": 8},
        )
    except Exception:
        # í´ë°±: í˜¸ì¶œ ì‹œ SystemMessageë¥¼ ê°™ì´ ì „ë‹¬
        res = agent.invoke(
            {"messages": [SystemMessage(content=SYSTEM_PROMPT),
                          HumanMessage(content=question)]},
            config={"recursion_limit": 8},
        )
    print(res["messages"][-1].content)

run_once("10ê³¼ 32ë¥¼ ë”í•œ ë‹¤ìŒ 2ë°°ë¡œ í•´ì¤˜.")
```

## 4) Agent ì‹¤í–‰ â€“ ìì—°ì–´ë¡œ Tool ì‚¬ìš© ì§€ì‹œ

ì•„ë˜ ì˜ˆì‹œ ì§ˆì˜ë¥¼ ì‹¤í–‰í•´ ë³´ê³ , ì–´ë–¤ Toolì„ ì„ íƒ/í˜¸ì¶œí–ˆëŠ”ì§€ í™•ì¸í•©ë‹ˆë‹¤.

### ì§ˆë¬¸ë³„ ë…ë¦½ ì‹¤í–‰(Stateless)

```python
questions = [
    "10ê³¼ 32ë¥¼ ë”í•œ ë‹¤ìŒ, ê·¸ ê²°ê³¼ë¥¼ 2ë°°ë¡œ ë§Œë“  ê°’ì„ ì•Œë ¤ì¤˜.",
    "'2025-11-13T06:00:00Z' ë¥¼ í•œêµ­ì‹œê°„ìœ¼ë¡œ ë°”ê¿”ì¤˜.",
    "7ê³¼ 8ì„ ê³±í•œ ë’¤, 10ì„ ë”í•´ì¤˜. í•„ìš”í•œ ê³„ì‚°ì€ ë„êµ¬ë¥¼ ì‚¬ìš©í•´.",
]

for q in questions:
    print("\n==============================")
    print("ì§ˆë¬¸:", q)
    res = agent.invoke(
        {"messages": [HumanMessage(content=q)]},
        config={"recursion_limit": 8},  # ë„êµ¬ í˜¸ì¶œ ë°˜ë³µ ì œí•œ
    )
    # ì‘ë‹µ ì¶”ì¶œ
    msgs = res.get("messages", [])
    answer = msgs[-1].content if msgs else res
    print("\nìµœì¢… ë‹µë³€:", answer)
```

### ëŒ€í™” ë§¥ë½ ìœ ì§€ (Stateful)

```python
from langchain_core.messages import HumanMessage, SystemMessage

history = []  # ì„¸ì…˜ íˆìŠ¤í† ë¦¬

def ask(q: str):
    print("\n==============================")
    print("ì§ˆë¬¸:", q)
    history.append(HumanMessage(content=q))
    out = agent.invoke({"messages": history}, config={"recursion_limit": 8})
    # LangGraph stateì—ëŠ” ì „ì²´ íˆìŠ¤í† ë¦¬ê°€ ë“¤ì–´ìˆìœ¼ë¯€ë¡œ, ìµœì‹ ìœ¼ë¡œ êµì²´
    history[:] = out["messages"]
    print("\nìµœì¢… ë‹µë³€:", history[-1].content)

for q in [
    "10ê³¼ 32ë¥¼ ë”í•œ ë‹¤ìŒ, ê·¸ ê²°ê³¼ë¥¼ 2ë°°ë¡œ ë§Œë“  ê°’ì„ ì•Œë ¤ì¤˜.",
    "'2025-11-13T06:00:00Z' ë¥¼ í•œêµ­ì‹œê°„ìœ¼ë¡œ ë°”ê¿”ì¤˜.",
    "7ê³¼ 8ì„ ê³±í•œ ë’¤, 10ì„ ë”í•´ì¤˜. í•„ìš”í•œ ê³„ì‚°ì€ ë„êµ¬ë¥¼ ì‚¬ìš©í•´.",
]:
    ask(q)
```
